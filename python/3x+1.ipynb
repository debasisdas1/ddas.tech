```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
fig = plt.figure(figsize = (10, 5))

final_list = []
x_range = list(range(1000,1001))
for x in x_range:
    arr = []
    val = 3*x + 1
    arr.append(val)
    while (val !=1):
        if val % 2 == 1:
            val = 3*val + 1
        else:
            val = val/2
        arr.append(val)
    final_list.append(arr)    

for item in final_list:
    plt.plot(item,marker='x')

plt.show()    
```
